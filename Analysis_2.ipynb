```python
import os
import pandas as pd

# Path to the folder containing CSV files
folder_path = r"C:\Users\Ab_moiz138\Desktop\Data Analysis\Realworld_DS_Problems\Sales_Analysis\Pandas-Data-Science-Tasks-master\SalesAnalysis\Sales_Data"

# List all .csv files in the directory
files = [file for file in os.listdir(folder_path) if file.endswith('.csv') and file != 'all_data.csv']

# Initialize an empty DataFrame to hold all the concatenated data
all_months_data = pd.DataFrame()

# Loop through the list of CSV files and concatenate them into one DataFrame
for file in files:
    file_path = os.path.join(folder_path, file)  # Join folder path and file name
    print(f"Reading file: {file_path}")  # Debug: Print the path of the file being read
    df = pd.read_csv(file_path)  # Read the CSV file
    all_months_data = pd.concat([all_months_data, df], ignore_index=True)  # Concatenate the data

# Export the final combined DataFrame to a CSV file

all_months_data.to_csv('all_data.csv', index=False)
print("Data concatenation complete. Saved to 'all_data.csv'.")
```

#### Question#4: What products are most often sold together:

```python
all_data.sample(50)
```

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
df = df.copy()
df.loc[:, 'Grouped Products'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head(3)
```

```python
df = df[['Order ID', 'Grouped Products']].drop_duplicates(keep = 'first') #remove duplicates by keeping 1 sales. 
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped Products']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)
```
